```python
# Import Dependencies
import pandas as pd
from pathlib import Path
```

```python
# Store file pathe in variable
file_path = Path("Resources/TobaccoUseData.csv")
```

```python
# Read datafile with the Pandas library
tobacco_data_df= pd.read_csv(file_path, encoding = "utf-8")
```

```python
# Clean the data and create the dataframe
clean_data_df = pd.DataFrame(tobacco_data_df[["YEAR","LocationDesc","TopicDesc","MeasureDesc","Response", "Data_Value", "Sample_Size", "Gender", "Race", "Age"]])
clean_data_df
```

```python
# Filter the Race column by "All Races"
clean_data_df = clean_data_df.loc[clean_data_df["Race"] == "All Races"]
clean_data_df
```

```python
# Reset the index
clean_data_df.reset_index(drop = True, inplace = True)
```
